```python
import os

pasta = r'C:\Users\brrochana1\Repositories\fesa\3sem\poo\n1\Exercícios\\'  # substitua pelo caminho da pasta que você deseja listar

for raiz, diretorios, arquivos in os.walk(pasta):
    for arquivo in arquivos:
        caminho_completo = os.path.join(raiz, arquivo)
        print(caminho_completo)
```

```python
palavra_antiga = 'Calculadora'  # substitua pela palavra que você deseja substituir
palavra_nova = 'Calculator'  # substitua pela palavra pela qual você deseja substituir a palavra antiga

for raiz, diretorios, arquivos in os.walk(pasta, topdown=False):
    for nome in arquivos:
        caminho_completo_antigo = os.path.join(raiz, nome)
        nome_novo = nome.replace(palavra_antiga, palavra_nova)
        caminho_completo_novo = os.path.join(raiz, nome_novo)
        os.rename(caminho_completo_antigo, caminho_completo_novo)
    for nome in diretorios:
        caminho_completo_antigo = os.path.join(raiz, nome)
        nome_novo = nome.replace(palavra_antiga, palavra_nova)
        caminho_completo_novo = os.path.join(raiz, nome_novo)
        os.rename(caminho_completo_antigo, caminho_completo_novo)
```
